# **Exploratory Data Analysis - Sample Super Store Dataset** 

```
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings 
```

```
#Importing dataset
sample_df = pd.read_csv("/content/SampleSuperstore.csv", error_bad_lines=False)
print("Data read successfully")
sample_df.head(10)
```

```
sample_df.shape
```

```
sample_df.drop(["Postal Code"],axis=1).describe()
```

```
sample_df.info()
```

```
sample_df.isnull().sum()
```

```
#Displaying duplicate values
sample_df.duplicated().sum()
```

```
#Removing duplicates
sample_df.drop_duplicates(subset=None, keep='first', inplace=True)
```

```
sample_df.duplicated().sum()
```

```
#Region v/s Count Data
sns.set()
sns.countplot(x = "Region", data = sample_df, palette = "Reds")
plt.show()
```

**We see that the West has the most business activity and engagemnet out of the 4 regions.**

```
#State v/s Count Data
plt.figure(figsize = (15,7))
sns.countplot(data = sample_df, x = 'State', order = sample_df.State.value_counts().index)
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Count")
plt.show()
```

**Out of the different states in the West region, California is the top state for business engagement.**

```
#Ship Mode v/s Count Data
sns.countplot(x = "Ship Mode", data = sample_df, palette = "Oranges")
plt.show()
```

**Here we see the most preferred "Ship Mode" which is Standard class**

```
#Category v/s Count Data
sns.countplot(x = 'Category', data = sample_df)
plt.show()
```

**In the Products category, we see that the no. of Office supplies is the category that is most sold out of the 3 categories.**

```
#Sub-Category v/s Count Data
sns.countplot(x = 'Sub-Category', data = sample_df)
plt.xticks(rotation=90)
plt.show()
```

**Further going into Sub-Category, we see that Binders, Paper and other office supplies are most sold.**

```
#Here we see that the Consumer Segment of Society is the largest customer segment
sns.countplot(x = "Segment", data = sample_df);
plt.show()
```

```
plt.figure(figsize=(15,8))
plt.hist(sample_df['Sales'],bins=(np.arange(0,500,10)), color = 'Red')
plt.title("Sales")
plt.xlabel("Sales")
plt.show()
```

**Here we see that the maximum sales made are approximately in the range (0-20)**

```
#Displaying Profits
plt.figure(figsize=(15,8))
plt.hist(sample_df['Profit'],bins=(np.arange(0,500,10)), color = 'Purple')
plt.title("Profits")
plt.xlabel("Profits Made")
plt.show()
```

**Here we see that the maximum profits made are approximately in the range (0-20)**

**By the above two graphs, we see that the Profits made and the number of Sales have a somewhat linear relationship.**

```
sales_data = sample_df.groupby(["State"],as_index=False)["Sales"].sum()
plt.figure(figsize=(20,8))
sns.barplot(sales_data["State"],sales_data["Sales"])
plt.xticks(rotation=90)
```

**Here we see the Sales made in each state. As spoken about earlier, California is at #1 and New York is at #2 having made the most sales.**

```
profits_data = sample_df.groupby(["State"],as_index=False)["Profit"].sum()
plt.figure(figsize=(20,8))
sns.barplot(profits_data["State"],profits_data["Profit"])
plt.xticks(rotation=90)
```

**Here we see the profitss made in each state. Since, California has made most sales, its profits are the most as well with New York close at #2.**

```
discount_data = sample_df.groupby(["State"],as_index=False)["Discount"].sum()
plt.figure(figsize=(20,8))
sns.barplot(discount_data["State"],discount_data["Discount"])
plt.xticks(rotation=90)
```

**Here, we see the discounts given by each state. We see that Texas gives the most discounts. By analysing earlier graphs, texas has made negative profit which means that it needs to cater to new sales methods and marketing techniques which do not include as many discounts as it is giving out now.**

```
#Displaying relationship data between Sales, Discount and Profit
plt.figure(figsize=(15,7))
sns.heatmap(sample_df.drop(['Postal Code','Quantity'],axis=1).corr())
print(sample_df.drop(['Postal Code','Quantity'],axis=1).corr())
```

**The purpose of the heatmap is to numerically display observations and relationship between the Sales, Discounts given and Profits made.**

**We see that the relationship between Sales and Discount is -0.02 (in negative) - which means they are inversely proportional but the value is very small, so it can be turned around by focusing on better discount rates in the respective areas which have made comparatively less (or negative) profits (such as Texas)**

**We see that relationship between Discount and Profit is -0.219 (negative) which shows that it is inversely proportional which is suggested in most states that have made negative (or comparatively less profits), discounts are given out a lot more than other states. Few examples of suchs states - Texas, Pennsylvania and Ohio.**

**We see that the relationship between Sales and Profit is 0.47 (positive) which shows a slight linear relationship, as discussed earlier.**
